## The ride sharing bonanza continues! Seeing the success of notable players like Uber and Lyft, you've decided to join a fledgling ride sharing company of your own. In your latest capacity, you'll be acting as Chief Data Strategist for the company. In this role, you'll be expected to offer data-backed guidance on new opportunities for market differentiation.

## You've since been given access to the company's complete recordset of rides. This contains information about every active driver and historic ride, including details like city, driver count, individual fares, and city type.

## Your objective is to estimate the relationship between four key variables:
Average Fare ($) Per City
Total Number of Rides Per City
Total Number of Drivers Per City
City Type (Urban, Suburban, Rural)

```python
import pandas as pd
```

```python
city= pd.read_csv('city_data.csv', low_memory=False)
rides=pd.read_csv('ride_data.csv', low_memory=False)
```

```python
city.head()
```

```python
rides.head()
```

## Total number of drivers per city without repitition of cities

```python
city.groupby(['city','type']).sum()
```

```python
city_group=city.groupby(['city','type']).sum()
```

```python
city_group.head()
```

```python
city_n=city_group.reset_index()
```

## Average Fare Per City

```python
rides.head()
```

```python
rides.shape
```

```python
rides.groupby('city').mean()
```

```python
rides[rides['city']=='Zimmermanmouth']['fare'].mean()
```

### The city Zimmermanmouth's mean matches with our answer above. Just Verifying!

```python
rides_n=rides.groupby('city').mean()
```

```python
rides_n.head()
```

```python
rides_n.drop('ride_id', axis=1, inplace=True)
```

```python
rides_n.reset_index()
rides_n.head()
```

## Total rides per city

```python
total_rides=rides.groupby('city').count()
```

```python
total_rides.head()
```

```python
total_rides=total_rides.drop(columns=['date','fare'])
```

```python
total_rides=total_rides.reset_index()
```

```python
total_rides.rename(columns={'ride_id':'rides'}, inplace=True)
```

```python
total_rides.head()
```

```python
total_rides.shape
```

## Combining all the data frames

```python
#city_group -> type, driver count per city
#rides_n -> mean fare per city
#total_rides -> total rides per city
x1=pd.merge(city_n, rides_n, on='city')
x=pd.merge(x1,total_rides, on='city')
x.drop('city', axis=1, inplace=True)
```

```python
x.head()
```

```python
x.columns=['City Type', 'Driver Count', 'Average Fare($)', 'Total Number of Rides (Per City)']
```

```python
x.head()
```

# End of Data Preparation

# Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import cycle
```

```python
plotData=x.copy()
```

```python
citytype = {}
colorcycle = cycle(['#87cefa', '#f08080', '#FFD700'])

for i in range(len(plotData['City Type'])):
    citytype[plotData['City Type'][i]] = next(colorcycle)

colors = [citytype[i] for i in plotData['City Type'] if i in citytype]
```

## Overall Distribution

```python
sns.set()
plotData.plot(kind='scatter', x='Total Number of Rides (Per City)', y='Average Fare($)', s=plotData['Driver Count']*20, c=colors, alpha=0.85, edgecolors="black", linewidth=0.9, figsize=(15,10), title='Ride Sharing Data Analysis')
plt.ylim(15, None)
plt.xlim(0, None)

#Creating a legend, that consists of 2 parts: City types and Driver count per city
for key in citytype:
    plt.scatter([],[], color=citytype[key], label=key, marker='o', edgecolor='white', linewidth=2)
for drcount in [5,25,45,70]:
    plt.scatter([], [], c='k', alpha=0.3, s=drcount*4, label=str(drcount) + ' drivers', edgecolor='black', linewidth=2)
plt.legend(scatterpoints=1, markerscale=2, loc=1, fontsize='large', labelspacing=1,
           title='City Types and\nDriver Count Per City:')
plt.savefig('Pyber Ride Sharing Data.png')
```

## Which city type is most expensive?

```python
from matplotlib import style
style.use('fivethirtyeight')

plotData2 = x.copy()
forpie = plotData2.groupby('City Type')
pie_df = forpie.sum()

# Find values < 20% for Explode
explode_f = [0.1 if (item/(pie_df['Average Fare($)']).sum()*100) < 20 else 0.0 for item in pie_df['Average Fare($)']]

farebysity = pie_df.plot(kind='pie', y= 'Average Fare($)', title=('% of Total Fares by City Type'),
                          colors=['#FFD700', '#87cefa', '#f08080'], autopct="%1.1f%%", shadow=True, startangle=360,
                          center=(0,0), explode = explode_f, figsize=(7,10), legend=False)
plt.ylabel('')
plt.tight_layout()
plt.savefig('Fares by City Type.png')
```

## How much for each city type?

```python
forpie.mean()
```

## Which city type has the maximum rides?

```python
explode_r = [0.1 if (item/(pie_df['Total Number of Rides (Per City)']).sum()*100) < 20 else
             0.0 for item in pie_df['Total Number of Rides (Per City)']]

farebysity = pie_df.plot(kind='pie', y= 'Total Number of Rides (Per City)', title=('% of Total Rides by City Type'),
                          colors=['#FFD700', '#87cefa', '#f08080'], autopct="%1.1f%%", shadow=True, startangle=360,
                          center=(0,0), explode = explode_r, figsize=(7,10), legend=False)
plt.ylabel('')
plt.tight_layout()
plt.savefig('Rides by City Type.png')
```

## How many drivers for each city type?

```python
explode_d = [0.1 if (item/(pie_df['Driver Count']).sum()*100) < 20 else 0.0 for item in pie_df['Driver Count']]

farebysity = pie_df.plot(kind='pie', y= 'Driver Count', title=('% of Total Drivers by City Type'),
                          colors=['#FFD700', '#87cefa', '#f08080'], autopct="%1.1f%%", shadow=True, startangle=360,
                          center=(0,0), explode = explode_d, figsize=(10,7), legend=False)
plt.ylabel('')
plt.tight_layout()
plt.savefig('Drivers by City Type.png')
```

# Findings: -
# Almost four - fifths (77,8%) of all drivers are from urban territories. They make 87% of their rides in cities and the rest in suburbs and rural areas. With such a high number, these rides rise only 46% of company's proceeds.
# Suburban rides make up only 26% of all rides, but the average fare is higher than that in cities, therefore they bring more than one third of a total revenue, that means that this area of activity can be developed in a future.
# Rural area fares are the most expensive ones - the average fare is almost 40% higher than the one in an urban city. With only 3% of all drivers and 5% of all rides, rural area brings more than 17% of a total revenue.
